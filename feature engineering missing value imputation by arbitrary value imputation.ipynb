### arbitarary imputation
this technique was derived from kaggle competition
it consists of replacing NaN by an arbitrary values

```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('titanic_train.csv',usecols=['Age','Fare','Survived'])
```

```python
df.head()
```

```python
def impute_nan(df,variable):
    df[variable+'_zeros']=df[variable].fillna(0)
    df[variable+'_hundred']=df[variable].fillna(100)
    
```

```python
df.Age.hist(bins=50)
```

```python
### Arbitrary values
### it should be more frequently present
```

```python
impute_nan(df,'Age')
```
